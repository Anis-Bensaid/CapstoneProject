```python
import pandas as pd
import numpy as np
import copy as cp

import statsmodels.api as sm
import matplotlib.pyplot as plt

import glob

import sklearn.metrics as metrics

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from heatmap import heatmap, corrplot
```

```python
onedrive_path = "C:/Users/cfowle/The Estée Lauder Companies Inc/TeamAnis - General/"
```

```python
def process_category(filename, target):
    df = pd.read_csv(filename)
    df = df.groupby(['sub_category', 'date']).sum()
    
    X = np.log(df[["topic_1", "topic_2", "topic_3", "topic_4"]] + 1)
    y = np.log(df[target] + 1)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle = False)
    
    lm = LinearRegression()
    lm.fit(X_train, y_train)
    pred = lm.predict(X_test)
    print("OSR^2:",  metrics.r2_score(pred, y_test))
    
    est = sm.OLS(y_train, X_train)
    est2 = est.fit()

    print(est2.summary())
```

```python
for file in glob.glob(onedrive_path + "Data/LDA Results/*.csv"):
    print(file)
    print("Sentiment Postiive")
    process_category(file, "sentiment_positive")
```
